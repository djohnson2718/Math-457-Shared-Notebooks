```python
def iterate(func, start_value, num_iter):
    x = start_value
    for i in range(1,num_iter+1):
        x = func(x)
        print i, ":", x
```

```python
def f(x):
    return x^2-1

def g(x):
    return -3/2*x^2 + 5/2*x +1
```

```python
iterate(f,.2, 100)
```

```python
iterate(g,0,20)
```

```python
iterate(g,0.0001,100)
```

```python
def iterate_until_tol(func, start_value, tol, target_value, max_iter):
        x = start_value
        for i in range(1,max_iter+1):
            x = func(x)
            print i, ":", x
            if abs(target_value - x) < tol:
                print "Mission accompished!", i, "iterations."
                break
        if i == max_iter:
            print "Mission failed, maximum iterations reached."
    
```

```python
iterate_until_tol(g,.5, .000001, 0, 100)
```
