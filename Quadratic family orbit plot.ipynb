```python
g = Graphics()
c=-2.3
while c < .5:
    c = c +.01
    x = 0
    for i in range(50):
        x = x^2 + c
        if i > 20: # give the orbit time to approach a fixed point
            if abs(x) <5: # through out the orbits that are approaching infinity 
                g = g + point((c,x))
        
plot(g)
```
