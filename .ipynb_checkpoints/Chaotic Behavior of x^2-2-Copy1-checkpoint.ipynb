```python
def f(x):
    return x^2-2
```

```python
def iterate(func, start, num):
    value = start
    for i in range(num):
        value = func(value)
        print i, ": ", value
```

```python
#0 is a fixed point
iterate(f,0,10)
```

```python
#if you move away from 0, even a little bit, you get "chaotic" behavior.
iterate(f,.0001,100)
```

```python
#fixed point
iterate(f, 1.0, 10)
```

```python
#chaotic
iterate(f, 0.5, 100)
```
